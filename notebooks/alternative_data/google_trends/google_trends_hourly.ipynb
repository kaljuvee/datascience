```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
data  = pytrends.get_historical_interest(["Tripadvisor Stock", 'TRIP stock price'], year_start=2020, month_start=11, day_start=20, hour_start=0, year_end=2020, 
                                 month_end=11, day_end=27, hour_end=0, cat=0, geo='', gprop='', sleep=0)
```

```python
data
```

```python
plt.figure(figsize=(20, 10))
plt.plot(data["Tripadvisor Stock"])
plt.plot(data["TRIP stock price"])
plt.legend(['Name', 'Ticker'])
```

```python
data.sort_values("TRIP stock price")
```

```python
data.index = data.index.strftime('%2020/%m/%d')
data.groupby('date').mean()
```

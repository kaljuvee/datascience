**Download historical price data**
*   [Yfinance](https://pypi.org/project/yfinance/)


```python
symbol = 'QQQ'
```

**Download Historical Price Data for Given Symbol**

```python
data = yf.download(symbol, interval = '1mo', start="2011-03-01", end="2020-08-01")
data.head()
```

```python
clean_df = data.dropna()
clean_df.head(20)
```

```python
clean_df.to_csv('data/price_history_qqq.csv')
```
